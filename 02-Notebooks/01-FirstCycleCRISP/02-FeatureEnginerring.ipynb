#### Functions

```python
def groupAge(num, breakpoints=[10, 20, 30, 45, 60, 70, 80, 120], result='01234567'):
    i = bisect.bisect(breakpoints, num-1)
    ageMapping = {
        0: 'Child',
        1: 'Teenager',
        2: 'Young',
        3: 'Adult',
        4: 'Midlife',
        5: 'Senior',
        6: 'Mature Adulthood',
        7: 'Late Adulthood'
    }
    
    return ageMapping[i]
```

# FEATURE ENGINEERING

```python
df02 = dfRaw.copy()
```

```python
# Group Age
df02['AgeGroup'] = df02['Age'].apply(lambda row: groupAge(row))


# Country Origin
originMap = {
    'France': 'Latin',
    'Spain': 'Latin',
    'Germany': 'Anglo-Saxons'
}

df02['Origin'] = df02['Geography'].map(originMap)
```
